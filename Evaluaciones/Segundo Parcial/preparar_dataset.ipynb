```python
# importamos librerias
import pandas as pd
import numpy as np
```

```python
def aumentar_ejemplos(archivo_original, archivo_aumentado, num_ejemplos_deseados):
    # Paso 1: Leer el archivo CSV original
    df_original = pd.read_csv(archivo_original)
    
    # Paso 2: Calcular la media de cada par de ejemplos
    df_medias = (df_original.shift() + df_original) / 2
    
    # Paso 3: Reemplazar los valores no finitos con un valor predeterminado
    df_medias = df_medias.replace([np.inf, -np.inf], np.nan).fillna(0)
    
    # Paso 4: Redondear y convertir a enteros
    df_medias = df_medias.round(decimals=0).astype(int)
    
    # Paso 5: Calcular la diferencia entre el número deseado de ejemplos y los ejemplos originales
    diff = num_ejemplos_deseados - len(df_medias)
    
    # Paso 6: Generar ejemplos adicionales seleccionando aleatoriamente los ejemplos originales
    if diff > 0:
        indices_muestreo = np.random.choice(df_medias.index, size=diff, replace=True)
        ejemplos_adicionales = df_medias.iloc[indices_muestreo]
        df_medias = pd.concat([df_medias, ejemplos_adicionales], ignore_index=True)
    
    # Paso 7: Guardar los nuevos ejemplos en un nuevo archivo CSV
    df_medias[:num_ejemplos_deseados].to_csv(archivo_aumentado, index=False)

# Uso de la función
aumentar_ejemplos('emails_pre.csv', 'emails_pre.csv', 10000)
```
