```python
df = datos
# Read the CSV file into a DataFrame
df = pd.read_csv('tortuga_pre.csv')

# Define the augmentation operations
# Here, we will add noise to the data
noise = np.random.normal(0, 0.1, df.shape)

# Apply the augmentation operations
df += noise

# Define the desired size of the example
desired_size = 50000

# Create a new DataFrame with the desired size
augmented_df = pd.DataFrame()

# While the size of the augmented DataFrame is less than the desired size,
# add new rows to the DataFrame
while len(augmented_df) < desired_size:
    # Randomly select a row from the original DataFrame
    row = df.sample(1)

    # Add the row to the augmented DataFrame
    augmented_df = augmented_df.append(row)

# Write the augmented DataFrame to a new CSV file
augmented_df.to_csv('tortuga_pre1.csv', index=False)
```

```python
# Obtener el índice de la columna
indice_columna = df.columns.get_loc('Age')

# Obtener el nombre de todas las columnas excepto la que se va a mover
columnas_sin_mover = [col for col in df.columns if col != 'Age']

# Reorganizar las columnas en el orden deseado
nuevo_orden_columnas = columnas_sin_mover + ['Age']
df = df[nuevo_orden_columnas]

# Guardar el DataFrame actualizado en un nuevo archivo CSV
df.to_csv('sleep_healt_pre.csv', index=False)
```

```python
df = pd.read_csv("datos_aumentados.csv")
# Reindexar el DataFrame
df = df.reset_index(drop=True)
# Calcular el factor de interpolación
factor_interpolacion = 10  # (200 ejemplos / 50 ejemplos)

# Crear una lista vacía para almacenar los DataFrames interpolados
df_interpolados = []

# Iterar sobre cada columna del DataFrame
for columna in df.columns:
    # Interpolar los valores de la columna y expandir el DataFrame
    columna_interpolada = df[columna].interpolate().repeat(factor_interpolacion)
    df_interpolados.append(columna_interpolada)

# Concatenar las columnas interpoladas en un nuevo DataFrame aumentado
df_aumentado = pd.concat(df_interpolados, axis=1)
```
