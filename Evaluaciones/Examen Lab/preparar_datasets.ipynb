```python
# agregando librerias a utilizar
import pandas as pd # manejo de dataset
import csv # leer archivos csv
import numpy as np # para calculos matematicos
```

```python
# cargar a una variable el dataset
df = pd.read_csv("datasets/dataset-tortuga.csv")
df
```

```python
# seleccionamos las columnas a eliminar
columnas_a_eliminar = ['NAME', 'USER_ID']
# eliminamos
df = df.drop(columnas_a_eliminar, axis=1)

# Guardar el DataFrame resultante en un nuevo archivo CSV
df.to_csv('tortuga_pre.csv', index=False)
```

```python
df
```

```python
# mapeo de la columna PROFILE
# - beginner_front_end =1
# - advanced_front_end =2
# - beginner_back_end =3
# - advanced_back_end =4
# - beginner_data_science =5
# - advanced_data_science =6
mapeo_profile = {'beginner_front_end': 1, 'advanced_front_end': 2, 'beginner_backend': 3, 'advanced_backend': 4, 'beginner_data_science': 5, 'advanced_data_science': 6}
# mapear la columna profile con los datos 
df['PROFILE'] = df['PROFILE'].map(mapeo_profile)
datos = df
# escriviendo el nuevo dataset
datos.to_csv('tortuga_pre.csv', index=False)
```

```python
# verificamos si existen otros valores de 1 al 6, las clases
set(datos['PROFILE'].tolist())
```

```python
# llenamos los espacios en nulo a cero
datos = datos.fillna(0)
# guardamos el archivo
datos.to_csv('tortuga_pre.csv', index=False)
```
